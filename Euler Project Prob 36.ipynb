```python
settings = [e for e in range(11,1000001)]
```

```python
settings
```

```python
palindrome = []
for i in settings:
    j = str(i)
    index = 0
    while True:
        if j[index] == j[len(j)-1-index]:
            index += 1
        elif j[index] != j[len(j)-1-index]:
            index = 0
            break
        
        if index == len(j):
            palindrome.append(i)
            break

palindrome
```

```python
f = []
for i in palindrome:
    j = bin(i)
    f.append(j)    
```

```python
f
```

```python
k = []
for i in f:
    j = i.replace('0b', '')
    k.append(j)
```

```python
k
```

```python
total = 0
for i in palindrome:
    total = total + int(i, 2)
```
