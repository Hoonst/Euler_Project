```python
def width(mat):
    maxim = 0
    for i in final:
        for j in range(0, len(i) - 3):
            maxx = i[j] * i[j +1] * i[j +2] * i[j + 3]
            if maxx > maxim:
                maxim = maxx
    return maxim
```

```python
def diagonal(mat):
    maxim = 0
    for i in range(0,17):
        for j in range(0,17):
            maxx = mat[i][j] * mat[i+1][j+1] * mat[i+2][j+2] * mat[i+3][j+3]
            if maxx > maxim:
                maxim = maxx
    return maxim
```
