```python
import numpy as np
```

```python
a = np.zeros((21,21))
```

```python
for i in range(len(a[0])):
    a[0][i] = 1
```

```python
for i in range(len(a[0])):
    a[i][0] = 1
```

```python
for i in a:
    for j in i:
        j = int(j)
```

```python
for i in range(1,21):
    for j in range(1,21):
        if a[i][j] == 0:
            a[i][j] = a[i-1][j] + a[i][j-1]
```
