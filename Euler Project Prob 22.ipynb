```python
name = open("prob 22.txt", "r")
```

```python
namess =name.read()
```

```python
namess = namess.replace('"', '')
namess = namess.split(',')
```

```python
namess.sort()
```

```python
namess[0][0]
```

```python
result = dict()
for i in namess:
    summ = 0
    for j in i:
        summ = summ + (ord(j)-64)
    result[i] = [summ]
    result[i].append(namess.index(i)+1)
    result[i].append(summ*(namess.index(i)+1))

result
```

```python
sum_of_all = 0

for i in result:
    sum_of_all += result[i][2]
    
sum_of_all
```
