9009같은 대칭수 만들기
두 자리 수를 곱해 만들 수 있는 대칭수 중 가장 큰 수는 9009

```python
setting = [(a,b) for a in range(100,1000) for b in range(100,1000)]
```

```python
setting
```

```python
settings = []
for i in setting:
    settings.append(i[0]* i[1])
    
settings
```

```python
palindrome = []
for i in settings:
    j = str(i)
    index = 0
    while True:
        if j[index] == j[len(j)-1-index]:
            index += 1
        elif j[index] != j[len(j)-1-index]:
            index = 0
            break
        
        if index == len(j):
            palindrome.append(i)
            break

palindrome
max(palindrome)
```
